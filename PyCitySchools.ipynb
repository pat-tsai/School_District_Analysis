```python
student_names = student_data_df['student_name'].tolist()
```

```python
faulty_names = []

for name in student_names:
 #   print(name.split(), len(name.split()))
    if len(name.split()) >= 3:
        faulty_names.append(name)

faulty_names
```

```python
len(faulty_names)
```

```python
prefixes, suffixes = [], []

for name in faulty_names: 
    if len(name.split()[0]) <= 4:
        prefixes.append(name.split()[0])
    elif len(name.split()[-1]) <= 3:
        suffixes.append(name.split()[-1])
```

```python
prefixes
```

```python
fixes_made = len(prefixes) + len(suffixes)
```

```python
fixes_made
```

```python
# returns unique values stored within list
set(prefixes)
```

```python
set(suffixes)
```

```python
# replacing prefixes/suffixes: .strip() will replace any character(s) that are within parenthesis, whereas .replace() will match exact cases
cleaned_prefixes = ['Dr.', 'Miss', 'Mrs.','Ms.','Mr.']
cleaned_faulty_list = []
for name in faulty_names:
    for prefix in cleaned_prefixes:
            if prefix == name.split()[0]:
                cleaned_faulty_list.append(name.replace(prefix, ""))
    if name.split()[-1] in suffixes:
        cleaned_faulty_list.append(name.replace(name.split()[-1],""))
```
